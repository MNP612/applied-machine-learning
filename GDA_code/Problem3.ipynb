```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def normal(mu,sig):
    sig_det = np.linalg.det(sig)
    sig_inv = np.linalg.inv(sig)
    N = np.sqrt((2*np.pi)**2 * sig_det)
    
    x=np.linspace(-7, 7, 60)
    y=np.linspace(-7, 7, 60)
    
    m = np.zeros([len(x),len(y)])
    
    for i in range(len(x)):
        for j in range(len(y)):
            pos = np.array([[x[i]],[y[j]]])-mu
            
            m[j,i] = (pos.T).dot(sig_inv).dot(pos)
            
    m = np.exp(-m/2)/N
    
    return x,y,m
```

* 1. where mu = [1,1] and sig = [[1,0],[0,2]]

```python
mu=np.array([1,1]).reshape(-1,1)
sig=np.array([[ 1 , 0], [0,  2]])

x,y,m = normal(mu, sig)

plt.axes()#.set_aspect('equal')
plt.contourf(x,y,m)
plt.colorbar()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, m,cmap='viridis', edgecolor='none')
plt.show()
```

```python
from scipy.stats import multivariate_normal

mu=[1, 1]
sig=[[1, 0], [0, 2]]
sig_det = np.linalg.det(sig)

x, y = np.mgrid[-7: 7:60j, -7: 7:60j]
pos = np.dstack((x, y))
rv = multivariate_normal(mu, sig)
m_test = rv.pdf(pos)

plt.axes().set_aspect('equal')
plt.contourf(x, y,m_test)
plt.colorbar()

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, m_test,cmap='viridis', edgecolor='none')
plt.show()
```

* 3. where mu1 = [0,2] and mu2 = [2,0] and sig = [[2,1],[1,2]]

```python
mu1=np.array([0,2]).reshape(-1,1)
mu2=np.array([2,0]).reshape(-1,1)
sig=np.array([[ 2 , 1], [1,  2]])

x,y,m1 = normal(mu1, sig)
x,y,m2 = normal(mu2, sig)

m = m1 - m2

plt.axes().set_aspect('equal')
plt.contourf(x,y,m)
plt.colorbar()

plt.show()
```
