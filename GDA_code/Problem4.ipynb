```python
import numpy as np
import random
from numpy import linalg as LA
from matplotlib import pyplot as plt
```

```python
def normal(x, mu, sig2):
    norm = (sig2*2*np.pi)**1/2*np.exp(-.5*(x-mu)**2/sig2)
    return norm
```

```python
np.random.seed(42)
X1 = np.random.normal(3, 3, 100)
X2 = 0.5*X1 + np.random.normal(4, 2, 100)

X = np.array((X1,X2))
```

```python
mean = np.array([sum(X[i])/len(X[i]) for i in range(2)])
mean
```

```python
cov = np.zeros((2,2))

for i in range(2):
    for j in range(2):
        cov[i,j] = sum((X[i]-mean[i])*(X[j]-mean[j]))/(len(X[i])-1)

cov
```

```python
value, vector = LA.eig(np.array(cov))
```
